```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
##SETENCES
sent = [
    "Hello World less",
    "Good Morning go hapy",
    "Python Programming no",
    "Machine Learning" ,
    "Artificial Intelligence",
    "Data Science"
]
```

```python
sent
```

```python
##DEFINE THE VOCAB SIZE
voc_size=10000
```

```python
##one hot encoder
onehot_repr = []
for words in sent:
    encoded_words = one_hot(words, voc_size)
    onehot_repr.append(encoded_words)

onehot_repr
```

```python
##word embedding represntation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np# we making our sent for same length 

sent_length=8

enbedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(enbedded_docs)

```

```python
#feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(enbedded_docs)
```
